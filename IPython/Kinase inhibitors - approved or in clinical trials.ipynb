# Kinase inhibitors - approved or in clinical trials

```python
import pandas as pd # Import pandas
import pylab
import numpy as np
import matplotlib.pyplot as plt
import rdkit.Chem as Chem
from rdkit.Chem import PandasTools
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole # Enables RDKit IPython integration
```

```python
dataframe = pd.read_csv('chembl_drugs-kinase_inh.txt', delimiter='\t')
```

```python
dataframe.columns
```

```python
dataframe[['DEVELOPMENT_PHASE', 'USAN_STEM_DEFINITION', 'USAN_STEM_SUBSTEM', 'ATC_CODE', 'ATC_CODE_DESCRIPTION', 'INDICATION_CLASS', 'DRUG_TYPE']].head()
```

Keep only nibs

```python
def hasnib(x):
    if 'nib' in str(x):
        return True
    else:
        return False
```

```python
dataframe['conatins_nib'] = dataframe['USAN_STEM_SUBSTEM'].map(hasnib)
```

```python
nibs = dataframe[dataframe['conatins_nib'] == True]
```

```python
nibs = nibs[nibs['CANONICAL_SMILES'].notnull()]
```

```python
PandasTools.AddMoleculeColumnToFrame(nibs, smilesCol='CANONICAL_SMILES')
```

```python
nibs = nibs[nibs['DRUG_TYPE'] != 'Oligonucleotide']
```

Get only INN name

```python
def innname(x):
    return x.split(' ')[0]
```

```python
nibs['INN_NAME'] = nibs['SYNONYMS'].map(innname)
```

```python
PandasTools.FrameToGridImage(nibs, legendsCol='INN_NAME', molsPerRow=7)
```

```python
nibs[['ROMol', 'SYNONYMS', 'RESEARCH_CODES', 'ATC_CODE', 'ATC_CODE_DESCRIPTION', 'INDICATION_CLASS', 'DEVELOPMENT_PHASE']].head(2)
```

```python
PandasTools.FrameToGridImage(nibs[nibs['DEVELOPMENT_PHASE'] == 4], legendsCol='INN_NAME', molsPerRow=4)
```

```python
len(nibs), len(nibs[nibs['DEVELOPMENT_PHASE'] == 4])
```

Save SMILES

```python
PandasTools.SaveSMILESFromFrame(nibs, 'nibs.smi', NamesCol='INN_NAME')
```

```python
PandasTools.SaveSMILESFromFrame(nibs, 'nibs_approved.smi', NamesCol='INN_NAME')
```
